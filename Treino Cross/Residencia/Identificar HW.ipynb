```python
import tensorflow as tf

print("Versão do TensorFlow:", tf.__version__)

gpus = tf.config.list_physical_devices('GPU')
if gpus:
    print(f"GPUs detectadas ({len(gpus)}):")
    for gpu in gpus:
        print(f" - {gpu}")
else:
    print("Nenhuma GPU detectada pelo TensorFlow.")
```
